# Notebook 01 - Inspección Inicial

En este notebook realizamos una primera exploración del dataset original 
sin modificar ningún dato. El objetivo es entender la estructura general, 
identificar problemas de calidad y formular preguntas de análisis.

```python
# Importamos las librerías necesarias para la inspección
import pandas as pd
import json
```

```python
# Cargamos el dataset original desde la carpeta raw (sin modificarlo)
with open('../data/raw/streaming_users_dirty.json') as f:
    data = json.load(f)

df = pd.DataFrame(data)
print("Dataset cargado correctamente.")
```

```python
# Verificamos cuántas filas y columnas tiene el dataset
print(f"El dataset tiene {df.shape[0]} filas y {df.shape[1]} columnas.")
```

```python
# Observamos los primeros registros para entender qué representa cada columna
df.head()
```

```python
# Revisamos el tipo de dato de cada columna
# Esto nos indica si las fechas, números y textos están bien clasificados
print("=== TIPOS DE DATOS ===")
print(df.dtypes)
```

```python
# Contamos cuántos valores faltantes hay en cada columna
print("=== VALORES NULOS POR COLUMNA ===")
print(df.isnull().sum())
print()
print(f"Total de valores nulos: {df.isnull().sum().sum()}")
```

```python
# Verificamos si hay filas duplicadas en el dataset
print(f"Filas duplicadas encontradas: {df.duplicated().sum()}")
```

```python
# Resumen estadístico de todas las columnas
# Nos permite detectar valores mínimos y máximos inusuales
df.describe(include='all')
```

```python
# Revisamos cuántos valores únicos tiene cada columna
# Útil para detectar inconsistencias en variables categóricas
for col in df.columns:
    print(f"{col}: {df[col].nunique()} valores únicos")
```

## Observaciones iniciales

Durante la inspección se identificaron los siguientes problemas de calidad
que deberán abordarse en la etapa de limpieza:

 **Valores nulos**: se detectaron nulos en `monthly_watch_time_mins` (193),
  `favorite_genre` (240) y `last_login_date` (320).

 **Filas duplicadas**: se encontraron 126 filas duplicadas.

 **Variables categóricas inconsistentes**: `subscription_plan`, `country` y
  `favorite_genre` presentan múltiples variantes para un mismo valor
  (mayúsculas, minúsculas, abreviaturas, errores tipográficos).

 **Valores imposibles**: `age` tiene valores negativos y superiores a 100.
  `customer_support_tickets` tiene valores negativos y outliers extremos (99, 150).

 **Fechas mal formateadas**: `last_login_date` está como texto y contiene
  fechas inválidas como "2026-15-03".

 **Outliers en variables numéricas**: `monthly_watch_time_mins` tiene valores
  negativos y un máximo de 99999, que no es realista.
