# Notebook 05 - Conclusiones

En este notebook se presentan las conclusiones del proyecto, respondiendo
a las preguntas de análisis definidas en la inspección inicial. Se diferencian
la evidencia obtenida, la interpretación realizada y la conclusión derivada.
Se incluyen también las limitaciones del trabajo y posibles mejoras futuras.

## Respuesta a las preguntas de análisis

### Pregunta 1
**¿El plan de suscripción influye en la cantidad de tiempo que un usuario
consume contenido mensualmente?**

Evidencia: el boxplot del EDA mostró distribuciones similares de tiempo
de visualización entre los tres planes de suscripción.

Interpretación: no se observaron diferencias marcadas entre planes,
lo que sugiere que el tiempo de consumo no depende del plan contratado.

Conclusión: el plan de suscripción no es un factor determinante del
tiempo de visualización mensual en este dataset.

### Pregunta 2
**¿Existe alguna preferencia de género según el país de origen del usuario?**

Evidencia: el análisis bivariado mostró variaciones en la distribución
de géneros favoritos entre países.

Interpretación: algunos países presentan preferencias más marcadas hacia
ciertos géneros, aunque las diferencias no son extremas.

Conclusión: el país de origen del usuario tiene una influencia moderada
en la preferencia de género, lo que podría orientar estrategias de
contenido por región.

### Pregunta 3
**¿A mayor edad, menor tiempo de visualización mensual?**

Evidencia: el mapa de correlaciones mostró una correlación muy baja
entre edad y tiempo de visualización.

Interpretación: la edad no presenta una relación lineal con el tiempo
de consumo, lo que indica que usuarios de distintas edades tienen
comportamientos de consumo similares.

Conclusión: la hipótesis de que los usuarios mayores consumen menos
contenido no se sostiene con la evidencia de este dataset.

### Pregunta 4
**¿Los usuarios que generan más tickets de soporte son los que menos
tiempo pasan en la plataforma?**

Evidencia: la correlación entre tickets de soporte y tiempo de
visualización fue cercana a cero en el mapa de calor del EDA.

Interpretación: los problemas técnicos que generan tickets no parecen
reducir el tiempo de uso de la plataforma de forma significativa.

Conclusión: no existe una relación lineal entre los tickets de soporte
y el tiempo de visualización, lo que sugiere que los usuarios continúan
consumiendo contenido a pesar de reportar incidencias.

## Hallazgos principales

- El dataset original presentó problemas significativos de calidad:
  duplicados, valores imposibles, inconsistencias en categóricas y nulos
  en múltiples columnas. Tras el proceso de limpieza se retuvo el 72.1%
  de los registros originales.
- Las variables numéricas del dataset no presentan correlaciones fuertes
  entre sí, lo que indica que cada una aporta información independiente.
- El plan de suscripción, la edad y los tickets de soporte no son
  predictores fuertes del tiempo de visualización mensual.
- El PCA confirmó que las tres variables numéricas contribuyen de forma
  equitativa a la varianza del dataset, sin que ninguna domine sobre las demás.

## Limitaciones

- El alcance de las conclusiones está condicionado por la calidad inicial
  del dataset y por las decisiones de limpieza documentadas en el Notebook 02.
  La eliminación del 27.9% de los registros puede haber afectado la
  representatividad de algunos grupos.
- Las variables disponibles son limitadas para explicar el comportamiento
  de consumo: no se cuenta con información sobre dispositivo, historial
  de contenido visto o frecuencia de acceso.
- Las correlaciones analizadas son lineales; podrían existir relaciones
  no lineales entre variables que este análisis no captura.

## Mejoras futuras

- Una mejora futura podría consistir en incorporar variables adicionales
  como frecuencia de acceso o tipo de contenido consumido, para ampliar
  el alcance del análisis de comportamiento de usuarios.
- Aplicar técnicas de segmentación como clustering permitiría identificar
  grupos de usuarios con perfiles similares, complementando los hallazgos
  del EDA y el PCA.
- Analizar la evolución temporal del comportamiento de los usuarios a
  partir de la variable `last_login_date` podría aportar información
  sobre patrones de retención en la plataforma.
