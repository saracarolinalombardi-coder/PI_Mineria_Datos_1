## Proceso de Limpieza y Preparación de Datos

El proceso de limpieza y preparación de datos fue fundamental para garantizar la calidad del análisis. Partiendo de un dataset con 8.160 registros y 8 columnas, se identificaron y resolvieron múltiples problemas de calidad:

Principales hallazgos y acciones tomadas:

Duplicados: Se eliminaron 126 filas duplicadas exactas que no aportaban información relevante.

Normalización de variables categóricas: Se unificaron 15 variantes del plan de suscripción, 26 variantes de país y 28 variantes de género favorito, reduciendo la inconsistencia en los datos categóricos.

Valores nulos: Se imputaron 193 valores nulos en tiempo de visualización (con la mediana), 240 en género favorito (con la moda) y 320 en fecha de último login (con la mediana de fechas válidas).

Valores imposibles: Se eliminaron registros con edades negativas o superiores a 100 años (74 registros), y valores negativos en tickets de soporte (27 registros) y tiempo de visualización (49 registros).

Outliers: Se aplicó winsorización a valores extremos en tiempo de visualización (límite superior de 2.704,7 minutos) y tickets de soporte (límite superior de 4 tickets), conservando así los registros sin distorsionar el análisis.

El resultado final fue un dataset limpio de 7.884 registros, con una retención del 96,6% de los datos originales y sin valores nulos en ninguna columna.



### Respuesta a las Preguntas de Análisis
1. ¿El plan de suscripción influye en la cantidad de horas que un usuario consume contenido mensualmente?

Respuesta: No se encontró evidencia de que el plan de suscripción influya significativamente en el tiempo de visualización. Los tres planes (Básico, Estándar y Premium) presentan distribuciones de tiempo de visualización muy similares, con medianas y rangos intercuartílicos prácticamente idénticos. Esto sugiere que el tipo de suscripción no es un factor determinante en el nivel de consumo de contenido, y que otros factores (como preferencias personales o hábitos de consumo) podrían tener mayor relevancia.

2. ¿Existe alguna preferencia de género según el país de origen del usuario?

Respuesta: El análisis de la variable favorite_genre después de la normalización muestra una distribución relativamente equilibrada entre los siete géneros principales, con "Comedia" como el género más frecuente (1.381 usuarios). Sin embargo, para determinar si existen preferencias específicas por país, sería necesario realizar un análisis de contingencia más detallado que relacione ambas variables categóricas. La limpieza realizada (normalización de 28 variantes a 7 géneros canónicos) permite ahora realizar este tipo de análisis cruzado con mayor precisión.

3. ¿A mayor edad, menor tiempo de visualización mensual?

Respuesta: El mapa de calor de correlaciones mostró una correlación muy baja (cercana a 0) entre la edad y el tiempo mensual de visualización, lo que indica que no existe una relación lineal entre ambas variables. La distribución de edades se concentra principalmente entre los 20 y 50 años, con mayor frecuencia en el rango de 25 a 40 años. Esto sugiere que la plataforma tiene una base de usuarios predominantemente adulta joven, pero dentro de ese rango no se observa una tendencia clara de disminución del consumo con la edad.

4. ¿Los usuarios que generan más tickets de soporte son los que menos tiempo pasan en la plataforma?

Respuesta: No se encontró evidencia que respalde esta relación. La correlación entre customer_support_tickets y monthly_watch_time_mins fue prácticamente nula en el mapa de calor. Tras la winsorización, los tickets de soporte se concentran mayoritariamente en valores bajos (0-2 tickets), y no se observa un patrón que vincule un mayor número de tickets con menor tiempo de visualización. Esto sugiere que los problemas de soporte no están directamente relacionados con el nivel de consumo de contenido.

### Hallazgos del Análisis de Componentes Principales (PCA)
El PCA aplicado a las variables numéricas (age, monthly_watch_time_mins y customer_support_tickets) reveló información importante sobre la estructura de los datos:

Varianza explicada: Los tres componentes principales explican aproximadamente un tercio de la varianza cada uno (33,75%, 33,22% y 33,03%), lo que indica que las variables no están correlacionadas entre sí y aportan información independiente.

Interpretación de las componentes:

PC1 (33,75%): Resume un perfil general del usuario, con contribuciones similares de las tres variables, destacando ligeramente los tickets de soporte y la edad.

PC2 (33,22%): Captura la diferencia entre usuarios mayores con menor tiempo de consumo frente a usuarios jóvenes con mayor tiempo de consumo.

PC3 (33,03%): Recoge la varianza restante, principalmente asociada a la edad.

Limitación importante: Con solo tres variables numéricas, la varianza se distribuye uniformemente entre los tres componentes, lo que impide una reducción significativa de dimensionalidad. Para superar el umbral del 80% de varianza acumulada, se necesitan las tres componentes originales. Esto confirma que las variables numéricas disponibles no son redundantes entre sí y que cada una aporta información única al análisis.
